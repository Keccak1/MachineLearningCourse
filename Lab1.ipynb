<h4>Importing data</h4>

```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_iris
```

```python
irisRaw = load_iris()
 
df = pd.DataFrame(data= np.c_[irisRaw['data'], irisRaw['target']],
                  columns= irisRaw['feature_names'] + ['target'])
```

<h5>Find the minimum and maximum petal length</h5>

```python
min_p_length = df['petal length (cm)'].min()
max_p_length = df['petal length (cm)'].max()

min_p_length_index = df['petal length (cm)'].idxmin()
max_p_length_index = df['petal length (cm)'].idxmax()

print(f"Min petal length {min_p_length} at index {min_p_length_index}")
print(f"Max petal length {max_p_length} at index {max_p_length_index}")
```

<h5>Calculate the standard deviation for each column</h5>

```python
for col in df.columns:
    print(f"Standart deviation of {col} is equal to {df[col].std()}")
```

<h5>Identify rows for which the length of the flower cup (sepal length) is greater than the average length (the whole set)</h5>

```python
df_filtered = df[df['sepal length (cm)'] > df['sepal length (cm)'].mean()]
```
